```python
from diffusers import StableDiffusionPipeline
import matplotlib.pyplot as plt
import torch
```

```python
image_model = "dreamlike-art/dreamlike-diffusion-1.0"

pipeline = StableDiffusionPipeline.from_pretrained(image_model, torch_dtype=torch.float16)
pipeline = pipeline.to("cuda")
```

```python
prompt = """ mass effect andromeda type astronaut with proper galaxy background the astronaut is male """
```

```python
image = pipeline(prompt).images[0]
plt.imshow(image)
plt.axis("off")
plt.show()
```

```python
images = pipeline(prompt,num_inference_steps=100, height = 512, width = 1024).images #more steps more better image accuracy since more times the model is trained
images[0]
```
